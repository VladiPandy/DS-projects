### Проверка расчета эффективности обогащения

Напишим функцию для рассчета эффективности обогащения для `rougher.output.recovery`:

```python
def efr(row):
    c=row['rougher.output.concentrate_au']
    f=row['rougher.input.feed_au']
    t=row['rougher.output.tail_au']
    recovery=((c*(f-t))/(f*(c-t)))*100
    return recovery   
```

Тоже самое сделаем для `final.output.recovery`:

```python
def eff(row):
    c=row['final.output.concentrate_au']
    f=row['rougher.input.feed_au']
    t=row['final.output.tail_au']
    recovery=((c*(f-t))/(f*(c-t)))*100
    return recovery   
```

Добавим столбецы с расчетной эффективностью:

```python
data_train['recovery_rougher']=data_train.apply(efr,axis=1)
```

```python
data_train['recovery_final']=data_train.apply(eff,axis=1)
```

Сравном визуально рассчетную и табличную:

```python
data_train[['recovery_rougher','rougher.output.recovery']].head()
```

```python
data_train[['recovery_final','final.output.recovery']].head()
```

Найдем MAE:

```python
mean_absolute_error(data_train['rougher.output.recovery'],data_train['recovery_rougher'])
```

```python
mean_absolute_error(data_train['final.output.recovery'],data_train['recovery_final'])
```

### Сравнения размера гранул

Посмотрим на сколько в среднем различаются размеры:

```python
(data_test['rougher.input.feed_size']-data_train['rougher.input.feed_size']).mean()
```

Теперь взглянем на распределение значений 

```python
d=data_test[['rougher.input.feed_size']].describe()
b=data_train[['rougher.input.feed_size']].describe()
pd.concat([d,b],axis=1)
```

```python
data_test['rougher.input.feed_size']=data_test['rougher.input.feed_size'].fillna(0)
```

Выдвеним нулевую гипотезу, что распределения гранул в выборках равны:

```python
alpha = 0.05

results = st.ttest_ind(
    data_train['rougher.input.feed_size'],
    data_test['rougher.input.feed_size'],
    equal_var=False)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Вероятность, что значение и больше окажется случайным при верной гипотезе, очень мала.

Для смещения среднего значения увеличим колличесвто строк с раззмером гранул от 30 до 55:

```python
def upsample(data_train, repeat):
    features_stay = data_train[data_train['rougher.input.feed_size']<30]
    features_up = data_train[(data_train['rougher.input.feed_size']<55)&(data_train['rougher.input.feed_size']>30)]
    features = data_train[data_train['rougher.input.feed_size']>=55]
    features_upsampled = pd.concat([features] + [features_stay] + [features_up] * repeat)
    features_upsampled = shuffle(
        features_upsampled, random_state=12345)
    
    return features_upsampled

data_train = upsample(data_train,2)
```

Посчитаем еще раз p-value:

```python
results = st.ttest_ind(
    data_train['rougher.input.feed_size'],
    data_test['rougher.input.feed_size'],
    equal_var=False)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Теперь вероятность , что значение и больше окажется случайным при верной гипотезе, равна 38%.

### Функции 

Напишем функцию для вычисления MAPE:

```python
def MAPE(actualy,predict):
    SUM=0
    for i in range(len(actualy)):
        if actualy[i] != 0 :
            MAPE=abs(actualy[i] - predict[i])/((abs(actualy[i]) + abs(predict[i]))/2)*100
            SUM+=MAPE
        else:
            SUM=SUM
    return (SUM/(len(actualy)))
```

<div style="background: #cceeaa; padding: 5px; border: 1px solid green; border-radius: 5px;">
<font color='green'> 
    <b><u>КОММЕНТАРИЙ РЕВЬЮЕРА</u></b>
<font color='green'><br>Такой есть вараинт ещё:

```python
def MAPE(y_true, y_pred):
    error = abs(y_true - y_pred)
    scale = (abs(y_true) + abs(y_pred)) / 2
    result = (error / scale).mean()
    return result
```

Теперь напишим функцию для вычисления sMAPE:

```python
def sMAPE(final_score_1,final_score_2):
    sMAPE=0.25*final_score_1+0.75*final_score_2
    return (print('Эффективность обогащения чернового концентрата:',final_score_1,'\n',
                 'Эффективность обогащения финального концентрата:',final_score_2,'\n',
                'Симметричное среднее абсолютное процентное отклонение:',sMAPE))
```

```python
my_scorer = make_scorer(MAPE, greater_is_better=False)
```

Сделаем функцию для кросс-валлидации:

```python
def cross_val(model,cv):
    scores_1=cross_val_score(model, features_rougher, target_rougher, cv=cv,scoring=my_scorer) 
    scores_2=cross_val_score(model, features_final, target_final, cv=cv,scoring=my_scorer) 
    final_score_1=sum(scores_1)/len(scores_1)
    final_score_2=sum(scores_2)/len(scores_2)
    score=sMAPE(final_score_1,final_score_2)
    return score
```
